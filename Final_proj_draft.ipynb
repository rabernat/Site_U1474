Holzkamper et al, 2009 - study of speleothem records from northern South Africa

They show growth at 58-46K and then at 40K which means that those were significantly wet periods

Goal of my project is to compare between: 
* speleothem records
* Ka/Ar ages 
* Physical Properties
* Astronomical cycles 


The end members of rivers are known: 
* Local South African rivers
* Limpopo
* Zambezi 

These river end members have some known geochemical characteristics. I hope that my combining these various records, I can confirm which end members are supplying more sediments during this time period. The speleothem record will serve as a check for which periods have greater moisture in south africa and how those periods show up in the other records.

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
```

```python

df = pd.read_csv("../Final_Project/DATA/NGR.csv")
df.head()
```

```python
df_n = df.rename(columns={"NGR total counts (cps)": "ngrT", "Depth CSF-A (m)": "depth"})
#df_n
```

```python

fig, ax =plt.subplots(figsize=(12,7))


df_n.plot(x ='depth', y='ngrT', ax=ax, c='k')

#ax.set_xlim(0, )
#ax.invert_yaxis()
ax.grid(True, which='major', linestyle='dotted')
#ax.set_ylim(5.5, 2.5)
ax.set_ylabel('Natural Gamma Ray')
ax.set_xlabel('Depth (m)')

#mpl.rcParams.update({'font.size': })
```
